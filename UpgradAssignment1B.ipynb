```python
#Plotting the growth of each states by the year with a best fit line
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(60, 20))
num = 1
d = dict()
for states in df_growth.columns[2:]:
    df_growth[states].dropna(inplace=True)
    plt.subplot(3,10,num)
    m,c = np.polyfit(df_growth[states].index, df_growth[states],1)
    plt.scatter(x = df_growth[states].index, y = df_growth[states])
    plt.plot(df_growth[states].index, m*df_growth[states].index+c,'--r')
    num = num + 1
    plt.title(states)
    d[states] = m
```

# We can compare the growth of these states by looking at the slope of the best fit line.


```python
#Slope of the best fit lines
print(d)
```

```python
Top_performing_states = sorted(d, key=d.get, reverse=True)[:3]
print(Top_performing_states)
#top 3 performing states are :['Goa', 'Manipur', 'Tripura']
```

```python

Lowest_performing_states = sorted(d, key=d.get, reverse=False)[:3]
print(Lowest_performing_states)

#Bottom 3 performing states are : ['Maharashtra', 'Maharashtra', 'Maharashtra']
```

#### Population Correlation with Primary, Upper Primary and Secondart Dropouts

```python
#Joining 1A data with data2 to include GSDP 
df_drpout = pd.merge(df_gdp_per, df_data1, how = 'inner', on = 'State')
df_drpout
```

```python
#Creating a new column for total dropouts for each states
df_drpout['Total'] = df_drpout['Primary - 2014-2015'] +  df_drpout['Upper Primary - 2014-2015'] +  df_drpout['Secondary - 2014-2015']
df_drpout
```

```python
plt.figure(figsize = (15,12))
n = 5
for i in range(1,n):
    plt.subplot(2,2,i)
    plt.scatter(df_drpout['GDP_per_capita'],df_drpout.iloc[:,1+i])
    plt.title('GSDP vs Drop out rate in '+df_drpout.columns[1+i])
    plt.ylabel('Percentage of Dropouts')
    plt.xlabel('GDP Per Capita')
```

```python
#As we see there there is a negative correlation Between drop out rate and GDP .
#GDP increases with decrease in Drop out rate. 
```
